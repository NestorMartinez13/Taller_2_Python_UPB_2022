```python
#Importar archivos csv y guardar sus datos en una lista vacía
import csv

dates=[]

sales=[]
expenses=[]

with open("sales.csv","r") as sales_csv:
    sales_reader = csv.DictReader(sales_csv, delimiter=";")
    for row in sales_reader:
        dates.append(row["date"].split("/")) #Se quita el caracter "/" y se crea una lista de listas [['1', '01', '2021'],...]
        sales.append(float(row["sales"])) #[779900.0, 200800.0, ...]

with open("expenses.csv", "r") as expenses_csv:
    expenses_reader = csv.DictReader(expenses_csv, delimiter=";")
    for row in expenses_reader:
        expenses.append(float(row["expenses"]))  #[635094.0, 994838.0,...]
```

```python
utilidades=[]
month_sales=[]        #Se crean tres listas vacias para separar por meses el total de ventas, gastos y utlidades
month_expenses=[]

Ene_acum=0
Ene_acum_sales=0      #Acumuladores donde se iran sumando los datos de cada dia
Ene_acum_expenses=0
for i in range(0,31):  #Se define el rango por cada mes dependiendo de los dias que tengan
    if dates[i][1] == "01": #Se separan por metodos de lista dependiendo del caracter del mes.
        Ene_acum_sales += sales[i]
        Ene_acum_expenses += expenses[i] #Se agregan los datos a los acumuladores
        Ene_acum +=sales[i]-expenses[i]
utilidades.append(Ene_acum)
month_sales.append(Ene_acum_sales)        #Finalmente se agregan como elementos a las listas, quedando alineadas con el
month_expenses.append(Ene_acum_expenses)  #numero y posicion de los meses

Feb_acum=0
Feb_acum_sales=0
Feb_acum_expenses=0
for i in range(31,59):
    if dates[i][1] == "02":
        Feb_acum_sales += sales[i]
        Feb_acum_expenses += expenses[i]
        Feb_acum += sales[i]-expenses[i]
utilidades.append(Feb_acum)
month_sales.append(Feb_acum_sales)
month_expenses.append(Feb_acum_expenses)

Mar_acum=0
Mar_acum_sales=0
Mar_acum_expenses=0
for i in range(59,90):
    if dates[i][1] == "03":
        Mar_acum_sales += sales[i]
        Mar_acum_expenses += expenses[i]
        Mar_acum += sales[i]-expenses[i]
utilidades.append(Mar_acum)
month_sales.append(Mar_acum_sales)
month_expenses.append(Mar_acum_expenses)

Abr_acum=0
Abr_acum_sales=0
Abr_acum_expenses=0
for i in range(90,120):
    if dates[i][1] == "04":
        Abr_acum_sales += sales[i]
        Abr_acum_expenses += expenses[i]
        Abr_acum += sales[i]-expenses[i]
utilidades.append(Abr_acum)
month_sales.append(Abr_acum_sales)
month_expenses.append(Abr_acum_expenses)

May_acum=0
May_acum_sales=0
May_acum_expenses=0
for i in range(120,151):
    if dates[i][1] == "05":
        May_acum_sales += sales[i]
        May_acum_expenses += expenses[i]
        May_acum += sales[i]-expenses[i]
utilidades.append(May_acum)
month_sales.append(May_acum_sales)
month_expenses.append(May_acum_expenses)

Jun_acum=0
Jun_acum_sales=0
Jun_acum_expenses=0
for i in range(151,181):
    if dates[i][1] == "06":
        Jun_acum_sales += sales[i]
        Jun_acum_expenses += expenses[i]
        Jun_acum += sales[i]-expenses[i]
utilidades.append(Jun_acum)
month_sales.append(Jun_acum_sales)
month_expenses.append(Jun_acum_expenses)

Jul_acum=0
Jul_acum_sales=0
Jul_acum_expenses=0
for i in range(182,212):
    if dates[i][1] == "07":
        Jul_acum_sales += sales[i]
        Jul_acum_expenses += expenses[i]
        Jul_acum += sales[i]-expenses[i]
utilidades.append(Jul_acum)
month_sales.append(Jul_acum_sales)
month_expenses.append(Jul_acum_expenses)

Ago_acum=0
Ago_acum_sales=0
Ago_acum_expenses=0
for i in range(211,243):
    if dates[i][1] == "08":
        Ago_acum_sales += sales[i]
        Ago_acum_expenses += expenses[i]
        Ago_acum += sales[i]-expenses[i]
utilidades.append(Ago_acum)
month_sales.append(Ago_acum_sales)
month_expenses.append(Ago_acum_expenses)

Sep_acum=0
Sep_acum_sales=0
Sep_acum_expenses=0
for i in range(243,273):
    if dates[i][1] == "09":
        Sep_acum_sales += sales[i]
        Sep_acum_expenses += expenses[i]
        Sep_acum += sales[i]-expenses[i]
utilidades.append(Sep_acum)
month_sales.append(Sep_acum_sales)
month_expenses.append(Sep_acum_expenses)

Oct_acum=0
Oct_acum_sales=0
Oct_acum_expenses=0
for i in range(273,304):
    if dates[i][1] == "10":
        Oct_acum_sales += sales[i]
        Oct_acum_expenses += expenses[i]
        Oct_acum += sales[i]-expenses[i]
utilidades.append(Oct_acum)
month_sales.append(Oct_acum_sales)
month_expenses.append(Oct_acum_expenses)

Nov_acum=0
Nov_acum_sales=0
Nov_acum_expenses=0
for i in range(304,334):
    if dates[i][1] == "11":
        Nov_acum_sales += sales[i]
        Nov_acum_expenses += expenses[i]
        Nov_acum += sales[i]-expenses[i]
utilidades.append(Nov_acum)
month_sales.append(Nov_acum_sales)
month_expenses.append(Nov_acum_expenses)

Dic_acum=0
Dic_acum_sales=0
Dic_acum_expenses=0
for i in range(334,365):
    if dates[i][1]:
        Dic_acum_sales += sales[i]
        Dic_acum_expenses += expenses[i]
        Dic_acum += sales[i]-expenses[i]
utilidades.append(Dic_acum)
month_sales.append(Dic_acum_sales)
month_expenses.append(Dic_acum_expenses)

#Se crea la lista con los meses
months = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"]
```

```python
import operator
import numpy as np
from datetime import datetime

data_sales = dict(zip(month_sales,months)) 
data_expenses = dict(zip(month_expenses, months)) #Se crean diccionarios donde el key es el valor de los gastos, ventas y utilidades
data_utilidades = dict(zip(utilidades, months))   #y los values son los meses

maxi_sales = max(data_sales.items(), key=operator.itemgetter(0)) #Se encuentra el maximo de las ventas con operator
mini_sales = min(data_sales.items(), key=operator.itemgetter(0))

maxi_expenses = max(data_expenses.items(), key=operator.itemgetter(0))
mini_expenses = min (data_expenses.items(), key=operator.itemgetter(0)) #Se encuentra el minimo de los gastos con operator

maxi_utilidades = max(data_utilidades.items(), key = operator.itemgetter(0))
mini_utilidades = min(data_utilidades.items(), key = operator.itemgetter(0)) 

total_sales = sum(month_sales)
total_expenses = sum(month_expenses) #Sumatoria de ventas, gastos y utilidades
total_utilidades = sum(utilidades)

prom_sales = int(total_sales/12)
prom_expenses = int(total_expenses/12)      #Promedio mensual de cada una
prom_utilidades = int(total_utilidades/12)

des_sales = np.std(month_sales)
des_expenses = np.std(month_expenses)      #Desviacion estandar de cada una utilizando std de numpy
des_utilidades = np.std(utilidades)

current_time = datetime.now()           #Fecha y hora
```

```python
#Se crea el .txt y se guarda con el nombre info.txt. Se escribe toda la información solicitada

with open("info.txt", "w") as info:
    info.write("\n\n INFORMACIÓN MENSUAL DE VENTAS, GASTOS Y UTILIDADES DE LA EMPRESA:\n\n\n")
    info.write(f"  -->Las maximas ventas se obtuvieron en {maxi_sales[1]}, con un total de: ${int(maxi_sales[0])}. Mientras que las menores ventas fueron en: {mini_sales[1]}, con un total de: ${int(mini_sales[0])}.\n\n")
    info.write(f"  -->Los mayores gastos fueron en {maxi_expenses[1]}, con un total de: ${int(maxi_expenses[0])}. Mientras que los menores gastos fueron en {mini_expenses[1]}, con un total de: ${int(mini_expenses[0])}.\n\n")
    info.write(f"  -->Las mayores utilidades se obtuvieron en {maxi_utilidades[1]}, con un total de: ${int(maxi_utilidades[0])}. Mientras que las menores utilidades se obtuvieron en {mini_utilidades[1]}, con un total de: ${int(mini_utilidades[0])}.\n\n\n\n")
    
    info.write(f"              -->Las ventas totales fueron: ${int(total_sales)}.                                -->En promedio en cada mes se vendió: ${prom_sales}.\n") 
    info.write(f"              -->Llos gastos totales fueron: ${int(total_expenses)}.                               -->En promedio en cada mes se gastó: ${prom_expenses}.\n")
    info.write(f"              -->Las utilidades sumaron un total de: ${int(total_utilidades)}.                       -->El promedio mensual de las utlidades fue de: ${prom_utilidades}.\n\n")
    
    info.write(f"              -->La desviación estandar para las ventas fue: {des_sales}.\n")
    info.write(f"              -->La desviación estandar para los gastos fue: {des_expenses}.\n")
    info.write(f"              -->La desviación estandar para las utilidades fue: {des_utilidades}.\n\n\n")
    
    info.write(f"Informe hecho por ---Nestor Martinez--- at {current_time}.")
```

```python
#Se procede a graficar las ventas, gastos y utilidades
import matplotlib as mpl
import matplotlib.pyplot as plt


fig, ax = plt.subplots(3,1, figsize=(15,15))

ax[0].plot(months,utilidades, marker="s")
ax[0].grid()
ax[0].set_title("Utilidades por meses")
ax[0].axis([0, 12, 0, max(utilidades)+2000000])
ax[0].set_xlabel("Meses")
ax[0].set_ylabel("valor en$ -- *10^7")
ax[0].legend(["utilidades"])

ax[1].plot(months,month_sales, marker="s")
ax[1].grid()
ax[1].set_title("Ventas por meses")
ax[1].axis([0, 12, 0, max(month_sales)+2000000])
ax[1].set_xlabel("Meses")
ax[1].set_ylabel("valor en $ -- *10^7")
ax[1].legend(["ventas"])

ax[2].plot(months,month_expenses, marker="s")
ax[2].grid()
ax[2].set_title("Gastos por meses")
ax[2].axis([0, 12, 0, max(utilidades)+2000000])
ax[2].set_xlabel("Meses")
ax[2].set_ylabel("valor en $ -- *10^7")
ax[2].legend(["gastos"])

plt.savefig("Graficas.png") #Guardado de las graficas con el nombre de "Graficas.png"
plt.show()
```
